## Find the Total Number of Students Having Marks Between 90 to 100 (inclusive) Using Between Method

```python
#without between
len(data[(data["Marks"]>=90) & (data["Marks"]<=100)])
```

```python
#using between
len(data[data["Marks"].between(90,100)])
```

## Find the Average Marks

```python
data["Marks"].mean()
```

## Apply Method

```python
def scale(x):
  return x/100
```

```python
data["scaled_marks"] = data["Marks"].apply(scale)
data.head(2)
```

```python
#using lambda
data['scaled_marks'] = data["Marks"].apply(lambda x: x/100)
data.head(1)
```

## Map Function

```python
data["encoded_gender"] = data["Gender"].map({"Male": 0, "Female": 1})
data.head(2)
```

## Drop the columns

```python
data.drop(["scaled_marks", "encoded_gender"], axis=1, inplace=True)
```

```python
data.sort_values(by="Marks", ascending=False, inplace=True)
data.head()
```
