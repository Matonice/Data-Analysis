```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/Salaries.csv")
df.head()
```

```python
df.columns
```

## Check null values in the dataset

```python
df.isnull().sum()
```

## Drop ID, Notes, Agency and Status Columns

```python
df.drop(['Id', 'Notes', 'Agency', 'Status'], axis=1, inplace=True)
```

## Get overall statistics about the dataframe

```python
df.describe(include="all")
```

## Find occurrence of the employee Names (Top 5)

```python
df['EmployeeName'].value_counts().head()
```

## Find the number of unique job titles

```python
df['JobTitle'].nunique()
```

## Total number of job titles contain captain

```python
sum(df['JobTitle'].str.contains("captain", case=False))
```

## Display all the employee names from fire department

```python
df[df['JobTitle'].str.contains('fire', case=False)]['EmployeeName']
```

## Find the minimum, maximum and average basepay

```python
df['BasePay'] = df['BasePay'].replace("Not Provided", np.nan)
df['BasePay'] = df['BasePay'].astype("float")
```

```python
df["BasePay"].describe()
```

## Replace "Not provided" in EmployeeName column with NaN


```python
df['EmployeeName'] = df['EmployeeName'].replace("Not Provided", np.nan)
```

## Drop the rows having 5 missing values

```python
df.drop(df[df.isnull().sum(axis=1)==5].index, axis=0)
```

## Find job title of Albert Pardini

```python
df[df['EmployeeName'] == "ALBERT PARDINI"]['JobTitle']
```

## How much does Albert Pardid make (including benefits)

```python
df[df['EmployeeName'] == "ALBERT PARDINI"]['TotalPayBenefits']
```

## Display the name of the person having the highest BasePay

```python
df[df['BasePay'] == df['BasePay'].max()]['EmployeeName']
```

## Find the average basepay of all employee per year

```python
df.groupby('Year')['BasePay'].mean()
```

## Find average basepay of all employee per job title

```python
df.groupby("JobTitle")['BasePay'].mean()
```

## Find average basepay of employee having job title accountant

```python
df[df['JobTitle'] == 'ACCOUNTANT']['BasePay'].mean()
```

```python
df['JobTitle'].value_counts().head()
```
